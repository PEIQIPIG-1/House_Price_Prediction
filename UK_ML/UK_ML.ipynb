```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_data = pd.read_csv("price_paid_records.csv")
```

```python
house_data
```

```python
sample_data = house_data.sample(n=100000, random_state=1)
sample_data.reset_index(drop=True, inplace=True)
```

```python
sample_data.info()
```

```python
sample_data['Date of Transfer'] = pd.to_datetime(sample_data['Date of Transfer'], format='%Y-%m-%d %H:%M', errors='coerce')
sample_data['Date of Transfer'] = sample_data['Date of Transfer'].dt.year
```

```python
sample_data
```

```python
sample_data = sample_data.rename(columns={'Date of Transfer':'Year'})
sample_data = sample_data.rename(columns={'Old/New':'New'})
```

```python
sample_data
```

```python
sample_data = sample_data.drop(['Transaction unique identifier', 'PPDCategory Type', 'Record Status - monthly file only'], axis=1)
```

```python
sample_data
```

```python
# Replace the abbreviation for Property Type
sample_data['Property Type'].replace('T', 'Terraced', inplace=True)
sample_data['Property Type'].replace('S', 'Semi-Detached', inplace=True)
sample_data['Property Type'].replace('D', 'Detached', inplace=True)
sample_data['Property Type'].replace('F', 'Flats/Maisonettes', inplace=True)
sample_data['Property Type'].replace('O', 'Other Property Type', inplace=True)

# Replace the abbreviation for New
sample_data['New'].replace('Y', True, inplace=True)
sample_data['New'].replace('N', False, inplace=True)


# Replace the abbreviation for Duration
sample_data['Duration'].replace('F', 'Freehold', inplace=True)
sample_data['Duration'].replace('L', 'Leasehold', inplace=True)
sample_data['Duration'].replace('O', 'Other Duration Type', inplace=True)

sample_data
```

```python
sample_data.info()
```

```python
categorical_features = ['Property Type', 'Duration', 'Town/City', 'District', 'County']
sample_data[categorical_features] = sample_data[categorical_features].astype('category')
sample_data.info()
```

```python
sample_data
```

```python
from sklearn.preprocessing import  OneHotEncoder

encoder = OneHotEncoder(sparse_output=False) 
one_hot_encoded = encoder.fit_transform(sample_data[categorical_features])
one_hot_encoded_columns = encoder.get_feature_names_out(categorical_features)
one_hot_encoded_df = pd.DataFrame(one_hot_encoded, columns=one_hot_encoded_columns)
sample_data_encoded = pd.concat([sample_data, one_hot_encoded_df], axis=1)
sample_data_encoded = sample_data_encoded.drop(columns=categorical_features)
sample_data_encoded
# sample_data = sample_data.join(pd.get_dummies(sample_data['Property Type'], prefix='Property Type')).drop(['Property Type'], axis=1)
# sample_data = sample_data.join(pd.get_dummies(sample_data['Duration'], prefix='Duration')).drop(['Duration'], axis=1)
# sample_data = sample_data.join(pd.get_dummies(sample_data['Town/City'], prefix='Town/City')).drop(['Town/City'], axis=1)
# sample_data = sample_data.join(pd.get_dummies(sample_data['District'], prefix='District')).drop(['District'], axis=1)
# sample_data = sample_data.join(pd.get_dummies(sample_data['County'], prefix='County')).drop(['County'], axis=1)
```

```python
sample_data_encoded.info()
```
