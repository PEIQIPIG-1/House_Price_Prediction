```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house_data = pd.read_csv("price_paid_records.csv")
```

```python
house_data
```

```python
house_data.info()
```

```python
house_data['Date of Transfer'] = pd.to_datetime(house_data['Date of Transfer'], format='%Y-%m-%d %H:%M', errors='coerce')
house_data['Date of Transfer'] = house_data['Date of Transfer'].dt.year
```

```python
house_data
```

```python
house_data = house_data.rename(columns={'Date of Transfer':'Year'})
house_data = house_data.rename(columns={'Old/New':'New'})
```

```python
house_data
```

```python
house_data = house_data.drop(['Transaction unique identifier', 'PPDCategory Type', 'Record Status - monthly file only'], axis=1)
```

```python
house_data
```

```python
# Replace the abbreviation for Property Type
house_data['Property Type'].replace('T', 'Terraced', inplace=True)
house_data['Property Type'].replace('S', 'Semi-Detached', inplace=True)
house_data['Property Type'].replace('D', 'Detached', inplace=True)
house_data['Property Type'].replace('F', 'Flats/Maisonettes', inplace=True)
house_data['Property Type'].replace('O', 'Other Property Type', inplace=True)

# Replace the abbreviation for New
house_data['New'].replace('Y', True, inplace=True)
house_data['New'].replace('N', False, inplace=True)


# Replace the abbreviation for Duration
house_data['Duration'].replace('F', 'Freehold', inplace=True)
house_data['Duration'].replace('L', 'Leasehold', inplace=True)
house_data['Duration'].replace('O', 'Other Duration Type', inplace=True)

house_data
```

```python
house_data.info()
```

```python
from sklearn.model_selection import train_test_split
x = house_data.drop(['Price'], axis=1)
y = house_data['Price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
train_data = x_train.join(y_train)
```

```python
train_data
```

```python
train_data = train_data.join(pd.get_dummies(train_data['Property Type'], prefix='Property Type')).drop(['Property Type'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data['Duration'])).drop(['Duration'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data['Town/City'], prefix='Town/City')).drop(['Town/City'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data['District'], prefix='District')).drop(['District'], axis=1)
train_data = train_data.join(pd.get_dummies(train_data['County'], prefix='County')).drop(['County'], axis=1)
train_data
```
