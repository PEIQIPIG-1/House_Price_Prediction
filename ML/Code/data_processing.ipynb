```python
from decimal import Decimal
# Convert house price to integer
data['price'] = data['price'].astype(float).astype(int)
data.head()
```

```python
# Combine to full address for coordinate translation
data['Address'] = data['street'].str.cat([data['city'],data['statezip'], data['country']],sep=' ')
data.head()
```

```python
#Creating Latitude and Longitude Attributes
data['Latitude'] = 0
data['Longitude'] = 0
data
```
